```python
def get_names_of_columns_with_type(d: pd.DataFrame, data_type):
    dict_columns = []
    for col in d.columns:
        for i in d[col]:
            if type(i) != data_type and type(i) == type(i):
                break
            elif type(i) == data_type:
                dict_columns.append(col)
                break
    return dict_columns
```

```python
get_names_of_columns_with_type(data, dict)
```

```python
dict_columns = get_names_of_columns_with_type(data, dict)
dict_columns_values = {}
for col in dict_columns:
    values = []
    for i in range(len(data[col])):
        values.append(data[col][i])
    dict_columns_values["{0}".format(col)] = values
```

```python
def get_vals_and_keys(series):
    keys = None
    vals = []
    for i in series:
        if type(i) == type(i):
            keys = list(i.keys())
            break
    for i in series:
        if i != i:
            a = []
            for j in range(0, len(keys)):
                a.append("NULL")
            vals.append(a)
        else:
            vals.append(list(i.values()))

    return [keys, vals]
```

```python
def create_dataframes(d_columns, d_columns_values):
    dataframes = {}
    for name in d_columns:
        d = get_vals_and_keys(d_columns_values["{0}".format(name)])
        dataframes["{0}".format(name)] = pd.DataFrame(d[1], columns = d[0])
    return dataframes
```

```python
dfs = create_dataframes(dict_columns, dict_columns_values)
dfs["musical_features"]
```

```python
def dict_columns_to_index_columns(main_df, column_name, secondary_df):
    main_df[column_name] = main_df[column_name].replace(list(main_df[column_name]), secondary_df.index)
    return main_df
```

```python
for name in dict_columns:
        data = dict_columns_to_index_columns(data, name, dfs[name])
```
